## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

quantidadeQuestoesMateria = {'mat':24,'por':18,'geo':8,'ing':8,'his':8,'fis':12,'qui':12}

acertosAlunoMat = (df_mat / quantidadeQuestoesMateria['mat'])*100

acertosAlunoPor = (df_por / quantidadeQuestoesMateria['por'])*100

acertosAlunoGeo = (df_geo / quantidadeQuestoesMateria['geo'])*100

acertosAlunoIng = (df_ing / quantidadeQuestoesMateria['ing'])*100

acertosAlunoHis = (df_his / quantidadeQuestoesMateria['his'])*100

acertosAlunoFis = (df_fis / quantidadeQuestoesMateria['fis'])*100

acertosAlunoQui = (df_qui / quantidadeQuestoesMateria['qui'])*100

````

````python
# 2) Seu código aqui
totalAcertosAluno = df_fis + df_geo + df_ing + df_mat + df_por + df_qui + df_his

totalAcertosAluno
````

````python
# 3) Seu código aqui

listaQuantidadeTotalQuestoes = pd.DataFrame.from_dict(quantidadeQuestoesMateria,orient='index')

listaQuantidadeTotalQuestoes.sum(0)

percentualGeralAluno = (totalAcertosAluno / int(listaQuantidadeTotalQuestoes.sum(0)))*100

percentualGeralAluno
````

````python
# 4) Seu código aqui
QuantidadeAlunosAcima45 = int(totalAcertosAluno[totalAcertosAluno['Qt_acertos'] > 45].count())
QuantidadeAlunosAcima45
````

````python
# seu código aqui

# desqualificados
zeroMat = df_mat[df_mat['Qt_acertos'] == 0]

zeroFis = df_fis[df_fis['Qt_acertos'] == 0]

zeroQui = df_qui[df_qui['Qt_acertos'] == 0]

desqualificados = zeroMat + zeroFis + zeroQui


desqualificados = pd.DataFrame({'Aluno':(zeroMat + zeroFis + zeroQui).index})

desqualificados
````

````python
# quantidade de desqualificados

qtdeDesqualificados = int(desqualificados.count())

print(f'Foram desqualificados {qtdeDesqualificados} alunos por zerarem em matemática, física ou química')
````

````python
# média em geografia e historia

hisGeo = pd.DataFrame({'Geografia':acertosAlunoGeo['Qt_acertos'],'História':acertosAlunoHis['Qt_acertos']})

desqualificadosHisGeo = desqualificados.join(hisGeo,on=None,how='left')

mediaGeoDesqualificados = desqualificadosHisGeo['Geografia'].mean()
mediaHisDesqualificados = desqualificadosHisGeo['História'].mean()

print(f'Os {qtdeDesqualificados} alunos desqualificados por zerarem em matemática, física e química tiveram média de {mediaGeoDesqualificados} % em Geografia e {mediaHisDesqualificados} % em história')
````
